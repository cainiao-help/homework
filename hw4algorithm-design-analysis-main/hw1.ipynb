1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9



idea:从一个数组中找出两个元素，使得后面的那个元素减去前面的元素最大,设置一个变量 min_price，记录从第 1 天到当前天的历史最低价格并记录此时的日期。只需要遍历一次，则时间复杂度为O(n)

````python
# add your code
#参考博客
def max_profit(profits):
    if not profits:
        return 0, 0, 0
    min_price = profits[0]
    max_profit = 0
    buy_day = 0
    sell_day = 0
    temp_buy_day = 0

    for i in range(1, len(profits)):
        if profits[i] < min_price:
            min_price = profits[i]
            temp_buy_day = i
        elif profits[i] - min_price > max_profit:
            max_profit = profits[i] - min_price
            buy_day = temp_buy_day
            sell_day = i

    return max_profit, buy_day, sell_day

profits = [3,2,1,-7,5,2,-1,3,-1]
result = max_profit(profits)
print("最大利润:", result[0])
print("买入日期:", result[1])
print("卖出日期:", result[2])
# your algorithm time complexity is :O(n)
````

2. \* 设计时间复杂度为 O(n)、空间复杂度为 O(1) 的算法解决此问题, 分析算法设计思路, 使用python编程实现

e.g.:
---
1. nums=[3,2,3], 返回3
2. nums=[2,2,1,1,1,2,2], 返回2


idea：由于多数元素的出现次数大于n/2,所以可以设置计数器count,若与候选元素相同则计数器加1，若不相同则减1，为0时则更新元素

````python
# add your code here
def majorityElement(nums):
    # 初始化候选元素和计数器
    candidate = nums[0]
    count = 1

    # 遍历数组
    for i in range(1, len(nums)):
        if nums[i] == candidate:
            # 当前元素与候选元素相同，计数器加 1
            count += 1
        else:
            # 当前元素与候选元素不同，计数器减 1
            count -= 1
            if count == 0:
                # 计数器为 0，更新候选元素并重置计数器
                candidate = nums[i]
                count = 1

    return candidate

# 测试示例
nums1 = [3, 2, 3]
print(majorityElement(nums1))  

nums2 = [2, 2, 1, 1, 1, 2, 2]
print(majorityElement(nums2))  
    
# your algorithm time complexity is :O(n)
````

1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1


idea：为了在O(log n)时间复杂度内找到数组中的局部最小值，我们可以利用二分查找法。该算法的核心思想是通过比较中间元素与其右侧元素的大小，逐步缩小搜索范围，直到找到一个局部最小值。初始化两个指针 left 和 right 分别指向数组的起始和末尾。在每次迭代中，计算中间位置 mid，比较 A[mid] 和 A[mid+1]如果 A[mid] > A[mid+1]，则右侧可能存在局部最小，将 left 移动到 mid+1,否则，左侧可能存在局部最小，将 right 移动到 mid，重复上述步骤直至 left 和 right 相遇，此时指向的元素即为局部最小值

````python
# add your code here
def find_local_min(arr):
    if len(arr) < 3:
        raise ValueError("数组必须有 3 个元素")
    left = 0
    right = len(arr) - 1
    while left < right:
        mid = (left + right) // 2
        if arr[mid] > arr[mid + 1]:
            left = mid + 1
        else:
            right = mid
    return arr[left]
    
arr1 = [10, 5, 2, 1, 7]
print(f"数组: {arr1}")
print(f"局部最小值: {find_local_min(arr1)}")
# your algorithm time complexity is :O(logn)
````

3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3


idea：1先快速排序算法，用于对数组进行排序，然后输出前k个最小值。快速排序的平均时间复杂度为 O(nlogn) 2使用堆方法找到并输出前k个最小的数。每次插入操作的时间复杂度为 O(logn),数组长度为 n.因此插入所有元素的总时间复杂度为时间复杂度O(nlogn).

````python
# add your code here
"""参考博客"""
"""1"""
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]  
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

def find_k_smallest_sort(arr, k):
    # 对数组进行排序
    sorted_arr = quick_sort(arr)
    # 获取前k个最小的数
    result = sorted_arr[:k]
    return result

arr=[5,4,3,2,6,1,88,33,22,107]
k = 5
print(f"前{k}个最小的数：{find_k_smallest_sort(arr, k)}")

"""2"""
"""时间复杂度为O(nlogn)"""

import heapq

def find_k_smallest_heap(arr, k):
    """
    使用堆方法找到并输出前k个最小的数。
    """
    # 创建一个堆
    heap = []
    # 将所有元素插入堆中
    for num in arr:
        heapq.heappush(heap, num)
    # 提取前k个最小的数
    result = [heapq.heappop(heap) for _ in range(k)]
    return result

arr=[5,4,3,2,6,1,88,33,22,107]
k = 5
print(f"前{k}个最小的数：{find_k_smallest_heap(arr, k)}")
# your algorithm time complexity is :O(nlogn)
````

举例1：数组A=["a", "da", "bde", "ab", "bc", "abdc", "cdba"], 排序后的数组应该为：['a', 'ab', 'abdc', 'bc', 'bde', 'cdba', 'da']

note：

-  两个字符之间的比较可以考虑比较他们对应的ASCII码值；
- python中可以使用`ord("a")`返回字符 “a”对应的ASCII值

idea:需要设计一个算法，能够在 O(n) 时间内对字符串数组进行排序，其中 n 是所有字符串的总字符数，步骤1：计算数组中所有字符串的长度，步骤2：将相同长度的字符串放在一组中，步骤3：在每个组内，按字典序对字符串进行排序，步骤4：合并结果

````python
# add your code here
def custom_sort(arr):
    # 按字符串长度分组
    length_dict = {}
    for s in arr:
        length = len(s)
        if length not in length_dict:
            length_dict[length] = []
        length_dict[length].append(s)
    
    # 对每个组内的字符串按字典序排序
    sorted_groups = []
    for length in sorted(length_dict.keys()):
        group = length_dict[length]
        # 对组内的字符串按字典序排序
        sorted_group = sorted(group)
        sorted_groups.extend(sorted_group)
    
    return sorted_groups

A1 = ["a", "da", "bde", "ab", "bc", "abdc", "cdba"]
A2 = ['ab', 'a', 'b', 'abc', 'ba', 'c']
A3 = ['aef', 'yzr', 'wr', 'ab', 'bhjc', 'lkabdc', 'pwcdba']

print(custom_sort(A1))
print(custom_sort(A2))
print(custom_sort(A3))
# your algorithm time complexity is :
````
